````python
import warnings
import sys, os, glob
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt

from PIL import Image
from tensorflow.keras import utils, callbacks
from keras_unet.losses import jaccard_distance
from keras_unet.metrics import iou, iou_thresholded # iou
````

````python
warnings.filterwarnings(action='ignore')
````

#### **모델 생성**

````python
from keras_unet.models import vanilla_unet

model = vanilla_unet(input_shape=(512, 512, 3))
utils.plot_model(model, show_shapes=True, dpi=50)
````

````python
from keras_unet.models import custom_unet

model = custom_unet(
    input_shape=(512, 512, 3),
    use_batch_norm=False,
    num_classes=1,
    filters=64,
    dropout=0.2,
    output_activation='sigmoid')
````

````python
utils.plot_model(model, show_shapes=True, dpi= 50)
````

## **데이터셋: Whale-tails**
---

### **| 데이터 처리**

#### **데이터 경로 확보**

````python
masks = glob.glob("D:Dataset/dataset/whales/*.png") # 고래 꼬리의 .png 확장자 데이터를 받아옴
orgs = list(map(lambda x: x.replace(".png", ".jpg"), masks)) # png -> jpg 이름을 바꾼 데이터
````

#### 이미지 크기 변환

````python
# PIL 의 Image 로 열어 이미지 크기 resize
imgs_list = []
masks_list = []
for image, mask in zip(orgs, masks):
    imgs_list.append(np.array(Image.open(image).resize((384,384))))
    masks_list.append(np.array(Image.open(mask).resize((384,384))))
````

#### **넘파이 변환**

````python
imgs_np = np.asarray(imgs_list)
masks_np = np.asarray(masks_list)
````

#### **데이터 확인**

````python
from keras_unet.utils import plot_imgs
plot_imgs(org_imgs=imgs_np, mask_imgs=masks_np, nm_img_to_plot=4, figsize=3)
````

#### **전처리**

````python
# size
print(imgs_np.max(), masks_np.max())
````

````python
# 255를 나눠 두 데이터의 scale 을 맞춤
x = np.asarray(imgs_np, dtype=np.float32)/255
y = np.asarray(masks_np, dtype=np.float32)
print(x.max(), y.max())
````

````python
print(imgs_np.shape, masks_np.shape)
````

````python
# masks_np의 채널에 대한 dims을 추가해서 두 데이터의 shape을 맞춤
masks_np = np.expand_dims(masks_np, 3)
imgs_np.shape, masks_np.shape
````

#### **Train / Test 데이터 분리**

````python
from sklearn.model_selection import train_test_split

x_train, x_val, y_train, y_val = train_test_split(x, y, test_size=0.9, random_state=0)

print("x_train: ", x_train.shape)
print("y_train: ", y_train.shape)
print("x_val: ", x_val.shape)
print("y_val: ", y_val.shape)
````

#### **이미지 증강**

````python
from keras_unet.utils import get_augmented

train_gen = get_augmented(
    x_train, y_train, batch_size=2,
    data_gen_args = dict(
        rotation_range=5.,
        width_shift_range=0.05,
        height_shift_range=0.05,
        shear_range=40,
        zoom_range=0.2,
        horizontal_flip=True,
        vertical_flip=False,
        fill_mode='constant'
    ))
````

````python
sample_batch = next(train_gen)
xx, yy = sample_batch
print(xx.shape, yy.shape)
plot_imgs(org_imgs=xx, mask_imgs=yy, nm_img_to_plot=2, figsize=3)
````

### **| 모델 생성 및 학습**

#### **모델 생성**

````python
input_shape = x_train[0].shape

model = custom_unet(
    input_shape,
    filters=32,
    use_batch_norm=True,
    dropout=0.3,
    dropout_change_per_layer=0.0,
    num_layers=4
)
````

````python
model.summary()
````

#### **컴파일 및 학습**

````python
# 체크포인트
model_filename = 'model/segm_model_v3.h5'
callback_checkpoint = callbacks.ModelCheckpoint(
    model_filename, 
    verbose=1, 
    monitor='val_loss', 
    save_best_only=True,
)
````

````python
# 컴파일
model.compile(
    optimizer= tf.keras.optimizers.Adam(), 
    #optimizer=SGD(lr=0.01, momentum=0.99),
    loss='binary_crossentropy',
    #loss=jaccard_distance,
    metrics=[iou, iou_thresholded]
)
````

````python
history = model.fit(
    train_gen,
    steps_per_epoch=200,
    epochs=50,
    validation_data=(x_val, y_val),
    callbacks=[callback_checkpoint]
)
````

#### **Plot training history**

````python
from keras_unet.utils import plot_segm_history
plot_segm_history(history)
````

#### **Plot original + ground truth + pred + overlay (pred on top of original)**

````python
model.load_weights(model_filename)
y_pred = model.predict(x_val)
````

````python
plot_imgs(org_imgs=x_val, mask_imgs=y_val, pred_imgs=y_pred, nm_img_to_plot=10, figsize=3)
````

## 데이터셋: **kz-isbi-challenge**
---

### **데이터 처리**

#### **데이터 경로 확보**

````python
masks = glob.glob("D:Dataset/dataset/isbi2015/train/label/*.png")
orgs = glob.glob("D:Dataset/dataset/isbi2015/train/image/*.png")
````

#### **이미지 크기 변환**

````python
imgs_list = []
masks_list = []
for image, mask in zip(orgs, masks):
    imgs_list.append(np.array(Image.open(image).resize((512,512))))
    
    im = Image.open(mask).resize((512,512))
    #width, height = im.size   # Get dimensions
    #left = (width - 388)/2
    #top = (height - 388)/2
    #right = (width + 388)/2
    #bottom = (height + 388)/2
    #im_cropped = im.crop((left, top, right, bottom))
    masks_list.append(np.array(im))
````

#### **넘파이 변환**

````python
imgs_np = np.asarray(imgs_list)
masks_np = np.asarray(masks_list)
````

#### **데이터 확인**

````python
print(imgs_np.shape, masks_np.shape)
````

````python
plot_imgs(org_imgs=imgs_np, mask_imgs=masks_np, nm_img_to_plot=10, figsize=3)
````

#### **전처리**

````python
# scale 조절
print(imgs_np.max(), masks_np.max())
````

````python
x = np.asarray(imgs_np, dtype=np.float32)/255
y = np.asarray(masks_np, dtype=np.float32)/255
print(x.max(), y.max())
````

````python
# dims 추가
print(x.shape, y.shape)
````

````python
x = np.expand_dims(x, 3)
y = np.expand_dims(y, 3)
print(x.shape, y.shape)
````

#### **Train / Test 데이터 분리**

````python
x_train, x_val, y_train, y_val = train_test_split(x, y, test_size=0.5, random_state=0)

print("x_train: ", x_train.shape)
print("y_train: ", y_train.shape)
print("x_val: ", x_val.shape)
print("y_val: ", y_val.shape)
````

#### **이미지 증강**

````python
train_gen = get_augmented(
    x_train, y_train, batch_size=1,
    data_gen_args = dict(
        rotation_range=15.,
        width_shift_range=0.05,
        height_shift_range=0.05,
        shear_range=50,
        zoom_range=0.2,
        horizontal_flip=True,
        vertical_flip=True,
        fill_mode='constant'
    ))
val_gen = get_augmented(
    x_val, y_val, batch_size=1)
````

````python
sample_batch = next(train_gen)
xx, yy = sample_batch
print(xx.shape, yy.shape)
plot_imgs(org_imgs=xx, mask_imgs=yy, nm_img_to_plot=2, figsize=3)
````

### **모델 생성 및 학습**

#### **모델 생성**

````python
input_shape = x_train[0].shape

model = custom_unet(
    input_shape,
    use_batch_norm=False,
    num_classes=1,
    filters=64,
    dropout=0.2,
    output_activation='sigmoid'
)
````

````python
model.summary()
````

````python
utils.plot_model(model, to_file='model.png', dpi=50)
````

#### **컴파일 및 학습**

````python
model_filename = 'model/segm_model_v0.h5'
callback_checkpoint = callbacks.ModelCheckpoint(
    model_filename, 
    verbose=1, 
    monitor='val_loss', 
    save_best_only=True,
)
````

````python
model.compile(
    optimizer=tf.keras.optimizers.SGD(lr=0.01, momentum=0.99),
    loss='binary_crossentropy',
    metrics=[iou, iou_thresholded]
)
````

````python
history = model.fit(
    train_gen,
    steps_per_epoch=200,
    epochs=10,  
    validation_data=val_gen,
    validation_steps=15,
    callbacks=[callback_checkpoint]
)
````

````python
history2 = model.fit(
    train_gen,
    steps_per_epoch=200,
    initial_epoch=10,
    epochs=20,  
    validation_data=val_gen,
    validation_steps=15,
    callbacks=[callback_checkpoint]
)
````

````python
history3 = model.fit(
    train_gen,
    steps_per_epoch=200,
    initial_epoch=20,
    epochs=30,  
    validation_data=val_gen,
    validation_steps=15,
    callbacks=[callback_checkpoint]
)
````

````python
loss = []
val_loss = []
iou = []
val_iou = []

loss += history.history["loss"]
val_loss += history.history["val_loss"]
iou += history.history["iou"]
val_iou += history.history["val_iou"]

loss += history2.history["loss"]
val_loss += history2.history["val_loss"]
iou += history2.history["iou"]
val_iou += history2.history["val_iou"]

loss += history3.history["loss"]
val_loss += history3.history["val_loss"]
iou += history3.history["iou"]
val_iou += history3.history["val_iou"]

epochs = np.arange(30)+1
````

````python
plt.figure(figsize=(6,4))
plt.plot(epochs, loss, label="loss", lw=3, color="r")
plt.plot(epochs, val_loss, label="val_loss", lw=3, color="b")
plt.xlabel("Epochs", fontsize=15);plt.ylabel("Loss",fontsize=15);
plt.xticks(fontsize=12);plt.yticks(fontsize=12)
plt.legend(fontsize=12);
````

````python
plt.figure(figsize=(6,4))
plt.plot(epochs, iou, label="IoU", lw=3, color="r")
plt.plot(epochs, val_iou, label="val_IoU", lw=3, color="b")
plt.xlabel("Epochs", fontsize=15);plt.ylabel("IoU",fontsize=15);
plt.xticks(fontsize=12);plt.yticks(fontsize=12)
plt.legend(fontsize=12);
````
