````python
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams['font.size'] = 15
plt.rcParams['xtick.labelsize'] = 8
plt.rcParams['ytick.labelsize'] = 8
plt.rcParams['figure.figsize'] = (8,5)
# 칼라 맵
plt.rcParams["image.cmap"] = "rainbow"
````

### 일변수 스칼라 함수
---

하나의 인풋(=스칼라, 일변수), 하나의 아웃풋(=스칼라, 스칼라함수)

\begin{align}
y = f(x)
\end{align}

#### **예제 01**

\begin{align}
f(x) = x^2+2x+1
\end{align}

* 함수 생성하기

````python
def f1(x):
    return x**2 + 2*x + 1
````

* 인풋 정의 (-10~10, 100개)

````python
xx = np.linspace(-10,10,100)
fx = f1(xx)

xx.shape, fx.shape
````

* 2d플롯으로 시각화

````python
f, ax = plt.subplots()
ax.plot(xx, fx)
ax.set(title = "f(x) according to x", xlabel="x", ylabel="f(x)")

x = 5
ax.plot(x, f1(x), marker="o", c="r") # 특정 x 에 따른 f(x)
ax.text(x-1.3, f1(x)+5, f"({x},{f1(x)})")
````

* 그래프 설명
  * x가 가질수 있는 값 : 모든 실수
  * f(x)가 가질 수 있는 값: 0 이상 모든 실수
  * 그래프의 특징 : x가 -1일때 최소값을 가지고 -1에서 멀어질 때 아웃풋이 커진다.

### 다변수 스칼라 함수
---

여러개의 인풋(=다변수), 하나의 아웃풋(=스칼라, 스칼라함수)


\begin{align}
y = f(𝐱)\\
\end{align}



#### **예제 03**

\begin{align}
f(x, y) = x+y
\end{align}

* 함수 생성하기

````python
def f3(x,y):
    return x+y
````

* 인풋 정의 (x, y, 모두 -15~20 값)

````python
xx = np.arange(-15,25) # 벡터x
yy = np.arange(-15,25) # 벡터y
X, Y = np.meshgrid(xx, yy)
f3(X,Y)
````

* 3d플롯으로 시각화

````python
# 3차원 플롯 생성
fig = plt.figure(figsize=(12,10))
ax = plt.axes(projection="3d")

# x, y, z 축 데이터 입력
ax.scatter3D(X,Y, f3(X,Y), marker=".", color="k")
surf = ax.plot_surface(X,Y, f3(X,Y), cmap="rainbow", alpha=0.5)

# 컬러바
fig.colorbar(surf,
             label="f(x,y)",
             shrink=0.6,
             location="top",
             aspect=30)

# 설명
ax.set(title="f(x,y) according to x,y", xlabel="x", ylabel="y", zlabel="f(x,y)")
````

* 그래프 설명
  * x, y가 가질수 있는 값 (인풋) : 
  * f(x)가 가질 수 있는 값 (아웃풋) : 
  * 그래프의 특징 :

#### **예제 04**

\begin{align}
f(x, y) = (x-5)^2+(y-5)^2 
\end{align}

* 함수 생성하기

````python
def f4(x,y):
    return (x-5)**2 + (y-5)**2 
````

* 인풋 정의(x, y 모두 -15~20 사이 값)

````python
xx = np.linspace(-15,20,100)
yy = np.linspace(-15,20,100)

X , Y = np.meshgrid(xx,yy)
f4xy = f4(X,Y)
````

* 3d플롯으로 시각화

````python
# 3차원 플롯 생성
fig = plt.figure(figsize=(12,10))
ax = plt.axes(projection="3d")

# x, y, z 축 데이터 입력
surf = ax.plot_surface(X,Y, f4xy, cmap="rainbow", alpha=0.5)

# 컬러바
fig.colorbar(surf,
             label="f(x,y)",
             shrink=0.6,
             location="top",
             aspect=30)

ax.set(title="f(x,y) according to x,y", xlabel="x", ylabel="y", zlabel="f(x,y)")
````

* 그래프 설명
  * x, y가 가질수 있는 값 (인풋) : x,y : 모든 실수
  * f(x)가 가질 수 있는 값 (아웃풋) : f(x,y) 0 보다큰 모든 실수
  * 그래프의 특징 : (x,y)=(5,5) 에서 최소값 0 을 가지고 5,5에서 멀어질수록 값이 커진다.

#### **예제 05**

\begin{align}
f(x_1, x_2) = 50(x_2-x_1^2)^2+(2-x_1)^2
\end{align}

* 함수 생성하기

````python
def f5(x1,x2):
    return 50*((x2-x1**2)**2)+(2-x1)**2
````

* 인풋 정의 (x: -2 ~ 2 20개, y: -1~3 20개)

````python
xx = np.linspace(-2,2, 200)
yy = np.linspace(-1,3, 200)

X, Y = np.meshgrid(xx,yy)

f5xy = f5(X,Y)
````

* 3d플롯으로 시각화

````python
# 3차원 플롯 생성
fig = plt.figure(figsize=(12,10))
ax = plt.axes(projection="3d")

# x, y, z 축 데이터 입력
surf = ax.plot_surface(X,Y, f5xy, cmap="rainbow", alpha=0.5)

# 컬러바
fig.colorbar(surf,
             label="f(x,y)",
             shrink=0.6,
             location="top",
             aspect=30)

ax.set(title="f(x,y) according to x,y", xlabel="x", ylabel="y", zlabel="f(x,y)")
````

* $x_1$, $x_2$의 범위를 둘다 -4부터 4로 변경하여 다시 그려보세요.

````python
xx = np.linspace(-4,4, 200)
yy = np.linspace(-4,4, 200)

X, Y = np.meshgrid(xx,yy)

f5xy = f5(X,Y)

# 3차원 플롯 생성
fig = plt.figure(figsize=(12,10))
ax = plt.axes(projection="3d")

# x, y, z 축 데이터 입력
surf = ax.plot_surface(X,Y, f5xy, cmap="rainbow", alpha=0.5)

# 컬러바
fig.colorbar(surf,
             label="f(x,y)",
             shrink=0.6,
             location="top",
             aspect=30)

ax.set(title="f(x,y) according to x,y", xlabel="x", ylabel="y", zlabel="f(x,y)")
````

* 그래프 설명
  * x, y가 가질수 있는 값 (인풋) : x,y 는 모든 실수
  * f(x)가 가질 수 있는 값 (아웃풋) : f(x,y)는 0보다 큰 모든실수
  * 그래프의 특징 : (2,4)에서 최소값 0 을 가지고 abs(x),y 값이 커질수록 f(x,y)값이 커진다.

## 일변수 벡터함수
---

하나의 인풋(=스칼라, 일변수), 여러개의 아웃풋(=스칼라, 스칼라함수)




\begin{align}
f(x) = (f_1(x), f_2(x), ..., f_n(x))^T
\end{align}

#### **예제 06**

\begin{align}
f(x) = \begin{pmatrix}
h(x)\\ 
g(x)\\ 
\end{pmatrix}  = \begin{pmatrix}
x^2\\ 
5x\\ 
\end{pmatrix}
\end{align}

* 함수 $f(x)$ 생성하기

````python
def f6(x):
    return np.array([x**3, np.cos(x)])
````

* 인풋 정의 (-10~10 20개)

````python
xx = np.linspace(-10,10,200)
hx, gx = f6(xx)
print(xx.shape, hx.shape, gx.shape)
````

* 2d플롯으로 시각화

````python
f, ax = plt.subplots(figsize=(8,6))
# ax.plot(hx, gx)
ax.set(title = "f(x) space", xlabel="h(x)", ylabel="g(x)")

scat = ax.scatter(hx,gx, c=xx, cmap="rainbow")

fig.colorbar(scat)
````

* 그래프 설명
  * x가 가질수 있는 값 (인풋) : 모든 실수
  * g(x)가 가질 수 있는 값 (첫번째 아웃풋) : -1<=g(x)<=1 인 모든 실수
  * h(x)가 가질수 있는 값 (두번째 아웃풋) : 모든 실수

#### **예제 07**

\begin{align}
f(t) = \begin{pmatrix}
x(t)\\ 
y(t)\\ 
z(t)\\ \end{pmatrix} = \begin{pmatrix}
sin(6t)\\
\frac{t}{4}\\
\frac{t^2}{2}\\
\end{pmatrix}
\end{align}

* 함수 생성하기

````python
def f7(t):
    return np.array([np.sin(6*t), t/4, (t**2)/2])
````

* 인풋 정의 (-1~3 200개)

````python
t = np.linspace(-1,3,200)

A, B, C = f7(t)
````

* 3d플롯으로 시각화

````python
# 3차원 플롯 생성
fig = plt.figure(figsize=(12,10))
ax = plt.axes(projection="3d")

# x, y, z 축 데이터 입력
surf = ax.scatter(A ,B, C)


ax.set(title="X,Y,Z according to t", xlabel="x(t) : $sin(6t)$", ylabel="y(t) : $t/4$", zlabel="z(t) : $t^2/2$")
````

* 그래프 설명
  * t가 가질수 있는 값 (인풋) : 모든실수
  * x(t)가 가질 수 있는 값 (첫번째 아웃풋) :  -1~1 사이 모든실수
  * y(t)가 가질 수 있는 값 (두번째 아웃풋) :  모든실수
  * z(t)가 가질 수 있는 값 (세번째 아웃풋) :  0보다큰 모든실수

#### **예제 08**

\begin{align}
\tau(t) = \begin{pmatrix}
x(t)\\ 
y(t)\\ 
z(t)\\ \end{pmatrix} = \begin{pmatrix}
cos(t)\\
sin(t)\\
t\\
\end{pmatrix}
\end{align}

* 함수 생성하기

````python
def f8(t):
    return np.array([np.cos(t), np.sin(t), t])
````

* 인풋 정의(-10 ~ 10 사이 200개)

````python
xx = np.linspace(-10,10, 200)
X, Y, Z = f8(xx)
````

* 3d플롯으로 시각화

````python
# 3차원 플롯 생성
fig = plt.figure(figsize=(12,10))
ax = plt.axes(projection="3d")

# x, y, z 축 데이터 입력
surf = ax.scatter(X , Y, Z)


ax.set(title="X,Y, according to t", xlabel="x(t) : $cos(t)$", ylabel="y(t) : $sin(t)$", zlabel="t")
````

* 그래프 설명
  * t가 가질수 있는 값 (인풋) : 모든 실수
  * cos(t)가 가질 수 있는 값 (첫번째 아웃풋) : -1~1 사이 모든 실수
  * sing(t)가 가질 수 있는 값 (두번째 아웃풋) : -1~1 사이 모든 실수

#### **예제 09**


\begin{align}
f(x,y) = \begin{pmatrix}
x^2\\ 
y^2\\ 
xy\\ \end{pmatrix}
\end{align}
