# 데이터 정제

## 칼럼은 index, year, m/d, day, time, weather, count 로 정한다.

```python
data01["year"] = data01.y
data01["count"] = data01.counts
data01.drop(["y","counts"], axis=1, inplace=True)
```

```python
data02["year"] = data02.y
data02["m/d"] = data02["m,d"]
data02["time"] = data02.t
data02["count"] = data02.counts
data02.drop(["y", "m,d", "t", "counts"], axis=1, inplace=True)
```

```python
data03["m/d"] = data03["m  d"]
data03["time"] = data03.t
data03.drop(["m  d", "t"], axis=1, inplace=True)
```

```python
data05["year"] = data05.yer
data05.drop(["yer"], axis=1, inplace=True)
```

```python
data = pd.concat([data01, data02, data03, data04, data05], axis=0, ignore_index=True)
```

## 결측값 확인

```python
data.isnull().sum()
```

## 데이터 값 기입 방식 변경

* 요일, 시간 ,날씨의 값을 문자열로 통일

```python
data = data.astype("str")
```

```python
data.head(3)
```

```python
data["day"] = data["day"].str.replace("0","sunday")
data["day"] = data["day"].str.replace("1","monday")
data["day"] = data["day"].str.replace("2","tuesday")
data["day"] = data["day"].str.replace("3","wednesday")
data["day"] = data["day"].str.replace("4","thursday")
data["day"] = data["day"].str.replace("5","friday")
data["day"] = data["day"].str.replace("6","saturday")
data["wheather"] = data["wheather"].str.replace("0","sunny")
data["wheather"] = data["wheather"].str.replace("1","blur")
data["wheather"] = data["wheather"].str.replace("2","rain")
data["time"] = data["time"].str.replace("0","morning")
data["time"] = data["time"].str.replace("1","noon")
data["time"] = data["time"].str.replace("2","evening")
data.head(5)
```

* 월과 요일 분리

```python
data["month"] = data["m/d"].str.split(" ", expand=True)[0]
data["days"] = data["m/d"].str.split(" ", expand=True)[1]
```

```python
data.head()
```

## 대여횟수 int

```python
data["count"] = data["count"].astype(float)
data["count"] = data["count"].round(0)
data = data.dropna()
data["count"] = data["count"].astype(int)
data.info()
```

## 확인

```python
print(data.day.unique())
print(data.time.unique())
print(data.wheather.unique())
print(data.month.unique())
print(data.days.unique())
```

```python
data.time = data.time.str.replace("monring", "morning")
```

```python
data.info()
```

* month, days int로 변환
* day를 dayofweek 으로 칼럼명 변환
* wheather 를 weather 로 칼럼명 변환

* 필요없는 칼럼 제거

```python
data["month"] = data["month"].astype(int)
data["days"] = data["days"].astype(int)

data["dayofweek"] = data["day"]
data["weather"] = data["wheather"]
data["index"] = data["Unnamed: 0"]
data = data.drop(["day", "wheather", "year", "m/d", "Unnamed: 0"], axis=1)
```

* 순서 정렬

```python
data = data[["index", "month", "days", "dayofweek", "time", "weather", "count"]]
```

# 데이터 분석

```python
# 시간대별 평균 대여량
data.groupby(["time"])["count"].mean().round(0)

# 한낮에 가장 빌리는 횟수가 적고
# 햇빛이 약한 아침, 저녁이 많음
```

## 대여수가 높거나 낮은날 특징찾기

```python
print(data["count"].describe())
```

```python
# 대여가 많이 된날 c
# 대여가 적게 된날 a
bins = [0,1112/3*1,1112/3*2,1112]
names = ['a','b','c']
data['count_a'] = pd.cut(data['count'], bins=bins, labels=names)
```

```python
# count_a 의 각각의 개수

# 1년중 평균적으로 아주 대여량이 높을때는 165/1112
# 평균적인 대여량이 낮다

data.count_a.value_counts()
```

```python
data.groupby(["weather"])["count_a"].count()
```

```python
pd.options.display.max_rows = 100
# 날씨와 대여량의 상관관계
data.groupby(["weather","count_a"])["time"].count().unstack()

# 흐린날의 높은 대여 확률은 29/302
# 비오는 날은 높은 100% 낮은 대여율
# 화창한 날은 136/573
```

```python
# 요일별 대여강도
data.groupby(["dayofweek","count_a"])["time"].count().unstack()
```

```python
# 시간대별 대여강도
data.groupby(["time","count_a"])["time"].count().unstack()
```

```python
# 월별 대여강도
print(data["month"].value_counts(sort=False))
data.groupby(["month","count_a"])["time"].count().unstack()

# 계절별 대여 횟수는 크게 차이가 없는것을 보인다.
# 킥보드는 날씨와 연관이 크고 계절에는 둔한것으로 보인다.
```

```python
# 날씨 / 요일별 대여강도
# 주말을 비교해본다.
data.groupby(["weather","dayofweek","count_a"])["count"].count().unstack()

# 흐린날은 평일에는 높은 대여량이 전혀 없지만 주말, 특히 일요일에는 중간, 높은 대여량을 보여준다.
# 비가 오는 날은 1년 전체에 걸쳐 평일, 주말 할 것 없이 낮은 대여량을 보여준다.
# 화창한 날에도 평일에는 높은 대여량을 보여주지 못한다. 하지만 화창한 주말에는 항상 높은 대여량을 보여준다.
# 주말이더라도 토요일과 일요일은 차이가 있다. 화창하든, 흐리든 일요일이 월등히 높은 대여량을 보여준다.a
```

```python
# 요일/ 시간 별 대여강도
data.groupby(["dayofweek","time","count_a"])["count"].count().unstack()

# 주말의 시간별을 비교해본다.
# 위에서 본바에 따르면 자정에는 아침과 저녁에 비해서는 저조했지만 주말의 경우에는 달랐다.
# 주말은 하루종일 꾸준히 대여량이 있고
```
