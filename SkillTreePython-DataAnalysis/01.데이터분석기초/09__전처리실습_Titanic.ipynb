11. 'Embarked' 필드의 결측치는 값이 가장 많은 'S'로 할당하기

12. 'Age' 필드의 결측치는 값이 중간값으로 할당하기

```python
df["Age"] = df["Age"].fillna(df["Age"].median())
```

```python
df["Age"].describe()
```

13. 'Name'필드에서 신분을 나타내는 단어를 뽑아서 'Title' 필드에 할당하기

```python
df
```

```python
# name 에서 , 를 기준으로 split
df_new = df["Name"].str.split(",", expand=True)
df_new
```

```python
# df_new 에서 다시 . 으로 split 을 해서 중간 단어를 뽑음
df_new_1 = df_new[1].str.split(".", expand=True)
df_new_1["Title"]=df_new_1[0]
```

```python
df_new_1
```

```python
df_1 = pd.concat([df, df_new, df_new_1], axis=1)
df = df_1.drop([0,1,2],axis=1)
```

```python
df.Title.value_counts()
# Jonkheer 작위없는 귀족
# mad·e·moi·selle = Mlle 프랑스 아가씨 Miss
# rev 목사, 신부
# col 해병, 해군 대령
# Capt 육군 대위, 해군 대령
# Don 신사
```

14. 'Title' 필드의 카테고리를 6개의 항목으로 변경하기

```python
# 6개 항목으로 통합
df['Title'] = df['Title'].replace(['Capt', 'Col', 'Major', 'Dr', 'Rev'], 'Officer')
df['Title'] = df['Title'].replace(['Jonkheer', 'Master'], 'Master')
df['Title'] = df['Title'].replace(['Don', 'Sir', 'the Countess', 'Lady', 'Dona'], 'Royalty')
df['Title'] = df['Title'].replace(['Mme', 'Ms', 'Mrs'], 'Mrs')
df['Title'] = df['Title'].replace(['Mlle', 'Miss'], 'Miss')
df['Title'] = df['Title'].replace(['Mr'], 'Mr')
```
