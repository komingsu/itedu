* 자료형 정리하기

```python
nums = ["displacement", "horsepower", "weight", "acceleration"] #연속형
noms = ["cylinders", "model_year", "origin"] # 이산 / 범주형

y = "mpg"
print("연속형자료", nums)
print("범주 / 이산자료", noms)
print("예측할 변수", y)
print("추가 전처리가 필요한 속성 : name")
```

* 데이터 정리하기
  * 데이터 소개 : 1970년 후반 ~ 1980년 초반의 mpg(자동차 연비)를 예측한다.
  * 자료형 소개
    * 연속 자료형
      * mpg : 연비
      * displacement: 배기량
      * horspower: 마력
      * acceleration : 가속도
    * 이산 자료형
      * cylinders: 실린더 수
      * weight: 무게
      * model_year: 모델 연도 
    * 범주 자료형
      * origin: 제조국가
      * name: 모델 이름
  * 모델링을 위한 자료형 구분
      * 연속


## 데이터 분포
---

### **| 데이터 분포 시각화**

* 연속 자료형 분포 그리기

```python
f, axes = plt.subplots(1,len(nums), figsize=(4*len(nums),4))
plt.subplots_adjust(wspace=0.2, hspace=0.2) # 좌우/ 상하 여백
axes = axes.flatten()
for col, ax in zip(nums, axes):
    sns.histplot(data=df, x=col, ax=ax).axvline(x=df[col].mean(),color="r")
plt.show()
```

* 이산 자료형 분포 그리기

```python
f, axes = plt.subplots(1,len(noms), figsize=(4*len(noms),4))
plt.subplots_adjust(wspace=0.3, hspace=0.2) # 좌우/ 상하 여백
axes = axes.flatten()
for col, ax in zip(noms, axes):
    sns.countplot(data=df, x=col, ax=ax)
plt.show()
```

### | **기술 통계량 시각화**

* 기술통계량 확인

* `boxplot` 으로 연속 자료형 기술 통계 시각화하기

```python
f, axes = plt.subplots(1,len(nums), figsize=(4*len(nums),4))
plt.subplots_adjust(wspace=0.2, hspace=0.2) # 좌우/ 상하 여백
axes = axes.flatten()
for col, ax in zip(nums, axes):
    sns.boxplot(data=df, x=col, ax=ax).axvline(x=df[col].mean(),color="r")
plt.show()
```

* `viloinplot` 으로 연속 자료형 기술 통계 시각화하기

```python
f, axes = plt.subplots(1,len(nums), figsize=(4*len(nums),4))
plt.subplots_adjust(wspace=0.2, hspace=0.2) # 좌우/ 상하 여백
axes = axes.flatten()
for col, ax in zip(nums, axes):
    sns.violinplot(data=df, x=col, ax=ax).axvline(x=df[col].mean(),color="r")
plt.show()
```

### **| 속성간 관계 시각화**

* `pairplot` 이용해 시각화 하기 

```python
# 시간이 아주 오개걸리기 때문에 소량의 샘플을 뽑아 그리는게 좋다.
df.sample(100)
sns.pairplot(df, hue="origin")
```

* `boxplot`으로 범주 속성별 예측값 관계 시각화하기

```python
f, axes = plt.subplots(1,len(noms), figsize=(6*len(noms),4))
plt.subplots_adjust(wspace=0.2, hspace=0.2) # 좌우/ 상하 여백
axes = axes.flatten()
for col, ax in zip(noms, axes):
    sns.boxplot(data=df,x=col, y="mpg", ax=ax)
```

* `scatterplot` 이용해 연속 자료형 별 예측값의 관계 시각화하기

```python
f, axes = plt.subplots(1,len(nums), figsize=(6*len(nums),4))
plt.subplots_adjust(wspace=0.2, hspace=0.2) # 좌우/ 상하 여백
axes = axes.flatten()
for col, ax in zip(nums, axes):
    sns.scatterplot(data=df,x=col, y=y, ax=ax)
```

* `hue` 옵션에 `origin` 추가해서 그리기 

```python
f, axes = plt.subplots(1,len(nums), figsize=(6*len(nums),4))
plt.subplots_adjust(wspace=0.2, hspace=0.2) # 좌우/ 상하 여백
axes = axes.flatten()
for col, ax in zip(nums, axes):
    sns.scatterplot(data=df,x=col, y=y, ax=ax, hue="origin")
```

* `regplot`으로 회귀 시각화

```python
f, axes = plt.subplots(1,len(nums), figsize=(6*len(nums),4))
plt.subplots_adjust(wspace=0.2, hspace=0.2) # 좌우/ 상하 여백
axes = axes.flatten()
for col, ax in zip(nums, axes):
    sns.regplot(data=df,x=col, y=y, ax=ax, scatter_kws={"s":10}, line_kws={"color":"r"})
```

* `residplot`으로 잔차 시각화
    * 회귀선이 x 축으로

```python
f, axes = plt.subplots(1,len(nums), figsize=(6*len(nums),4))
plt.subplots_adjust(wspace=0.2, hspace=0.2) # 좌우/ 상하 여백
axes = axes.flatten()
for col, ax in zip(nums, axes):
    sns.residplot(data=df,x=col, y=y, ax=ax, scatter_kws={"s":10})
```

* `lmplot`으로 `origin`별 회귀 시각화

```python
for col in nums:
    sns.lmplot(data=df,x=col, y=y, scatter_kws={"s":10}, hue="origin")
```
