## 데이터 분포
---

### **| 데이터 분포 시각화**

* 연속 자료형 분포 그리기

```python
f, axes = plt.subplots(1,len(nums), figsize=(4*len(nums),4))
plt.subplots_adjust(wspace=0.2, hspace=0.2) # 좌우/ 상하 여백
axes = axes.flatten()
for col, ax in zip(nums, axes):
    sns.histplot(data=df, x=col, ax=ax).axvline(x=df[col].mean(),color="r")
plt.show()
```

* 이산 자료형 분포 그리기

```python
f, axes = plt.subplots(3,3, figsize=(15,15))
plt.subplots_adjust(wspace=0.3, hspace=0.2) # 좌우/ 상하 여백
axes = axes.flatten()
for col, ax in zip(noms, axes):
    sns.countplot(data=df, x=col, ax=ax)
plt.show()
```

### | **기술 통계량 시각화**

* 기술통계량 확인

* boxplot 으로 연속 자료형 기술 통계 시각화하기

```python
f, axes = plt.subplots(1,len(nums), figsize=(4*len(nums),4))
plt.subplots_adjust(wspace=0.2, hspace=0.2) # 좌우/ 상하 여백
axes = axes.flatten()
for col, ax in zip(nums, axes):
    sns.boxplot(data=df, x=col, ax=ax).axvline(x=df[col].mean(),color="r")
plt.show()
```

* viloinplot 으로 연속 자료형 기술 통계 시각화하기

```python
f, axes = plt.subplots(1,len(nums), figsize=(4*len(nums),4))
plt.subplots_adjust(wspace=0.2, hspace=0.2) # 좌우/ 상하 여백
axes = axes.flatten()
for col, ax in zip(nums, axes):
    sns.violinplot(data=df, x=col, ax=ax).axvline(x=df[col].mean(),color="r")
plt.show()
```

### **| 속성간 관계 시각화**

* `pairplot` 이용해 시각화 하기 

```python
# pairplot 은 시간이 오래 걸리기 때문에 일부 샘플을 추출해 그리는게 좋습니다.
sns.pairplot(df, hue="survived")
```

* `displot` 이용해 연속 자료형 별 예측값의 분포 `kde` 선으로 시각화 하기 

```python
for col in nums:
      sns.displot(data=df, x=col, hue=y, kind='kde')
      plt.show()
```

* 위에서 그린 플롯에 `col` 옵션으로 `sex` 추가하기

```python
for col in nums:
    sns.displot(data=df, x=col, hue=y, kind='kde', col='sex')
    plt.show()
```

* 위에서 그린 플롯에 `hue` 옵션으로 `class` 추가하기

```python
for col in nums:
    sns.displot(data=df, x=col, hue="class", kind='kde', col='sex')
    plt.show()
```

* `age`과 `fare`의 산점도를 그리고 `hue`옵션에 예측값 추가하기

```python
sns.scatterplot(data=df,x="age", y="fare", hue=y)
```

* `lmplot`을 이용해 위의 산점도를 다시 그리고 `col` 옵션에 `class` 추가하기

```python
sns.lmplot(data=df,x="age", y="fare", hue=y, col="class")
```

* `histplot`으로 범주형 속성들과 예측값 히트맵 그리기

```python
f, axes = plt.subplots(3,3, figsize=(15,15))
axes = axes.flatten()
for col, ax in zip(noms, axes):
    sns.histplot(data=df, x=col, y=y,ax = ax)
plt.show()
```
