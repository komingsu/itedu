### | **데이터 확보**

데이터를 처리 하기 전 Step1 에서 진행 할 주제를 정의합니다. 

주제 : 분야별 재정구조 예산 추이 (지난 6~7년)

* 경로 설정

```python
data_paths = []

for file in glob.glob('재정데이터 시각화 경진대회/분야별 재정지출/*/*'):
    if "재정 구조" in file:
        data_paths.append(file)
    if "재정구조" in file:
        data_paths.append(file)
        
data_paths
```

```python
dfs = []
for data_path in data_paths:
    newcol=data_path.split("\\")[1]
    data_path = data_path.replace('\\', "\\\\") # \ 를 \\ 로 바꿔야 경로 설정가능
    df = pd.read_csv(data_path, encoding="CP949")
    df["분야"] = newcol
    dfs.append(df)
    
for data_path, df in zip(data_paths, dfs):
    print(data_path)
    print(df)
```

## 고용노동부 청년 관련 정책

```python
df_work = df_new[df_new.iloc[:,1]=='고용노동부']
df_young = df_work[df_work.loc[:,'단위사업명'].str.contains('청년')] ## 단위사업명 내에 청년이 포함된 경우에만 해당
```

청년과 관련된 정책 비중

```python
total_work = df_work.groupby(['회계년도'])['국회확정금액(천원)'].sum()
```

```python
subtotal_young = df_young.groupby(['회계년도'])['국회확정금액(천원)'].sum()
```

```python
young_ratio = pd.DataFrame(subtotal_young/total_work)
```

2019년부터 청년과 관련된 고용 노동부 사업의 비중이 증가하였습니다. 하지만 5%로 여전히 낮은 비중입니다.

```python
plt.bar(young_ratio.index.values,young_ratio.iloc[:,0])
plt.show()
```
