### **| 기본 사용**


| 표현 | 종류 |
|--|--|
|`-`  | Solid |
| `--` |Dashed|
|`:`|Dotted|
|`-.` |Dash-dot|

```
import matplotlib.pyplot as plt

plt.plot([1, 2, 3], [4, 4, 4], '-', color='C0', label='Solid')
plt.plot([1, 2, 3], [3, 3, 3], '--', color='C0', label='Dashed')
plt.plot([1, 2, 3], [2, 2, 2], ':', color='C0', label='Dotted')
plt.plot([1, 2, 3], [1, 1, 1], '-.', color='C0', label='Dash-dot')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.axis([0.8, 3.2, 0.5, 5.0])
plt.legend(loc='upper right', ncol=4)
plt.show()
```

````python
import matplotlib.pyplot as plt

plt.plot([1, 2, 3], [4, 4, 4], '-', color='C0', label='Solid')
plt.plot([1, 2, 3], [3, 3, 3], '--', color='C0', label='Dashed')
plt.plot([1, 2, 3], [2, 2, 2], ':', color='C0', label='Dotted')
plt.plot([1, 2, 3], [1, 1, 1], '-.', color='C0', label='Dash-dot')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.axis([0.8, 3.2, 0.5, 5.0])
plt.legend(loc='upper right', ncol=4)
plt.show()
````

### **| linestyle 지정하기**


| 표현 | 종류 |
|--|--|
|`solid`  | Solid |
| `dashed` |Dashed|
|`dotted`|Dotted|
|`dashdot` | Dash-dot|

```
plt.plot([1, 2, 3], [4, 4, 4], linestyle='solid', color='C0', label="'solid'")
plt.plot([1, 2, 3], [3, 3, 3], linestyle='dashed', color='C0', label="'dashed'")
plt.plot([1, 2, 3], [2, 2, 2], linestyle='dotted', color='C0', label="'dotted'")
plt.plot([1, 2, 3], [1, 1, 1], linestyle='dashdot', color='C0', label="'dashdot'")
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.axis([0.8, 3.2, 0.5, 5.0])
plt.legend(loc='upper right', ncol=4)
plt.tight_layout()
plt.show()
```

````python
plt.plot([1, 2, 3], [4, 4, 4], linestyle='solid', color='C0', label="'solid'")
plt.plot([1, 2, 3], [3, 3, 3], linestyle='dashed', color='C0', label="'dashed'")
plt.plot([1, 2, 3], [2, 2, 2], linestyle='dotted', color='C0', label="'dotted'")
plt.plot([1, 2, 3], [1, 1, 1], linestyle='dashdot', color='C0', label="'dashdot'")
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.axis([0.8, 3.2, 0.5, 5.0])
plt.legend(loc='upper right', ncol=4)
plt.tight_layout()
plt.show()
````

plt.plot([1, 2, 3], [2, 2, 2], linestyle='dashed', linewidth=10,
      dash_capstyle='butt', color='C1', label='dashed+butt')

      dash_capstyle='round', color='C1', label='dashed+round')

## **29. 이미지 저장하기**

### **| 기본사용**



```
import numpy as np
import matplotlib.pyplot as plt

x1 = np.linspace(0.0, 5.0)
x2 = np.linspace(0.0, 2.0)

y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)
y2 = np.cos(2 * np.pi * x2)

plt.subplot(2, 1, 1)                # nrows=2, ncols=1, index=1
plt.plot(x1, y1, 'o-')
plt.title('1st Graph')
plt.ylabel('Damped oscillation')

plt.subplot(2, 1, 2)                # nrows=2, ncols=1, index=2
plt.plot(x2, y2, '.-')
plt.title('2nd Graph')
plt.xlabel('time (s)')
plt.ylabel('Undamped')

plt.tight_layout()
# plt.show()
plt.savefig('savefig_default.png')
```

````python
import numpy as np
import matplotlib.pyplot as plt

x1 = np.linspace(0.0, 5.0)
x2 = np.linspace(0.0, 2.0)

y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)
y2 = np.cos(2 * np.pi * x2)

plt.subplot(2, 1, 1)                # nrows=2, ncols=1, index=1
plt.plot(x1, y1, 'o-')
plt.title('1st Graph')
plt.ylabel('Damped oscillation')

plt.subplot(2, 1, 2)                # nrows=2, ncols=1, index=2
plt.plot(x2, y2, '.-')
plt.title('2nd Graph')
plt.xlabel('time (s)')
plt.ylabel('Undamped')

plt.tight_layout()
plt.savefig('savefig_default.png')
````

````python
x1 = [1,2,3,4,5]
y1 = [1,2,4,8,10]
x2 = [1,2,3,4,5]
y2 = [10,5,4,2,1]

# 그래프
plt.plot(x2,y2,
         linestyle='dashed',
         linewidth=10,
         dash_capstyle='butt',
         color="cornflowerblue",
         label="Uncertainty")

plt.plot(x1,y1,
         marker="*",
         color="hotpink",
         markersize=20,
         label="Skill",
         linestyle='solid',
         linewidth=4,
         solid_capstyle='butt')

# 그리드
plt.grid(alpha=0.9, axis="y")

# 범례
plt.legend(loc="upper center")

# 타이틀
plt.title("My plot", fontsize=18,loc="left",pad=20)

# 레이블
plt.xlabel('X-Axis', fontsize=20)
plt.ylabel('Y-Axis', fontsize=20)

# 눈금
plt.tick_params(axis='x', direction='out', length=10, labelsize=15, width=5, labelcolor='darkcyan')
plt.tick_params(axis='y', direction='out', length=10, labelsize=15, width=5, labelcolor='darkcyan')

# 저장, Show
plt.tight_layout()
plt.savefig("ch0201_05")
````

####**예제 06**

본인 스타일 대로 데이터를 잘 나타내도록 그림을 그리고 `ch0201_06`으로 저장하세요.


````python
x1 = [1,2,3,4,5]
y1 = [1,2,4,8,10]
x2 = [1,2,3,4,5]
y2 = [10,5,4,2,1]

# 그래프
plt.plot(x2,y2,
         linestyle='dashed',
         marker="o",
         markersize=20,
         linewidth=10,
         dash_capstyle='butt',
         color="hotpink",
         label="Uncertainty")

plt.plot(x1,y1,
         marker="*",
         color="hotpink",
         markersize=20,
         label="Skill",
         linestyle='solid',
         linewidth=4,
         solid_capstyle='butt')

# 범례
plt.legend(ncol=2, loc="upper center", fontsize=13, frameon =True, shadow=True)

# 타이틀
plt.title("My hot ♥ pink plot", fontsize=30,pad=20, color="hotpink")

# 레이블
plt.xlabel('X-Axis', fontsize=20, fontdict={'color': 'hotpink', 'weight': 'bold', 'size': 'xx-large'})
plt.ylabel('Y-Axis', fontsize=20, fontdict={'color': 'hotpink', 'weight': 'bold', 'size': 'xx-large'})

# 눈금
plt.tick_params(axis='x', direction='out', length=10, labelsize=15, width=5, labelcolor='r')
plt.tick_params(axis='y', direction='out', length=10, labelsize=15, width=5, labelcolor='r')

# 저장, Show
plt.tight_layout()
plt.savefig("ch0201_06")
````
