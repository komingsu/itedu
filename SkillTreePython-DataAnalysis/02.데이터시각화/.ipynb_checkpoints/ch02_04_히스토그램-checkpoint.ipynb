### (1) `plt.hist`의 `histtypes` 옵션비교

```
x1 = np.random.normal(0, 1, 5000)
x2 = np.random.normal(0, 3, 5000)
x3 = np.random.normal(10, 3, 5000)

histtypes=['bar', 'barstacked', 'step', 'stepfilled']

```

````python
x1 = np.random.normal(0, 1, 5000)
x2 = np.random.normal(0, 3, 5000)
x3 = np.random.normal(10, 3, 5000)

histtypes=['bar', 'barstacked', 'step', 'stepfilled']

for i, histtype in enumerate(histtypes):
    plt.subplot(2,2, i+1)
    plt.hist([x1,x2,x3], histtype=histtype, label=["x1","x2","x3"], bins=50, alpha=0.7)
    plt.legend()
    plt.title(histtype)
plt.show()
````

### (2) 객체지향 사용하기


```
fig, ax = plt.subplots(2,2, figsize=size)
```

````python
fig, ax = plt.subplots(2,2)
ax = ax.flatten()

for i, histtype in enumerate(histtypes):
    ax[i].hist([x1,x2,x3], histtype=histtype, label=["x1","x2","x3"], bins=50, alpha=0.7)
    ax[i].legend()
    ax[i].set_title(histtype)
plt.show()
````

### (3) Iris 데이터 변수별 분포 그리기

```
           꽃받침 길이   꽃받침 너비   꽃의 길이  꽃의 너비
 샘플1      
 샘플2 
 ...
 샘플150
```


```
from sklearn.datasets import load_iris
iris = load_iris()
```


* `.feature_names` : 변수 이름
* `.target_names` : 타겟 이름
* `.data` : 데이터 
* `.target` : 타겟


````python
from sklearn.datasets import load_iris
iris = load_iris()
````

````python
x_names = iris.feature_names
y_names = iris.target_names
print("x names : ", x_names)
print("y names: ", y_names)

x_data = iris.data
y_data = iris.target
print("x_data" , x_data.shape)
print("y_data" , y_data.shape)
````

````python
plt.hist(y_data)
plt.xticks([0,1,2], y_names)
plt.title("Number of each target")
plt.show()
````

````python
plt.hist(x_data, histtype="stepfilled", bins=30, alpha=0.8, label=x_names)
plt.legend()
plt.show()
````

### 플러스과제

* 2 X 1 의 subplot 에  petal 끼리 / sepal 끼리 나눠서 그리기

````python
fig, ax = plt.subplots(2,1)
ax = ax.flatten()

sepal = x_data[:,:2]
sepal_name = x_names[:2]
ax[0].hist(sepal, histtype="stepfilled", bins=30, alpha=0.8, label=sepal_name)
ax[0].legend()
ax[0].set_title("sepal")

petal = x_data[:,2:]
petal_name = x_names[2:]
ax[1].hist(petal, histtype="stepfilled", bins=30, alpha=0.8, label=petal_name)
ax[1].legend()
ax[1].set_title("petal")

plt.show()
````

### (1)  지수분포(exponential distribution) 그리기




* 지수분포의 확률밀도 함수는 다음과 같습니다.


![](https://wikimedia.org/api/rest_v1/media/math/render/svg/7ae26c8e2d08abfda4102c31e1f380e7fa9068b7)

* 아래와 같이 넘파이 함수를 이용해 확률밀도 함수를 따르는 데이터셋을 얻을 수 있습니다.

```
np.random.exponential(scale=lambda, size=데이터개수)
```

* lambda 값이 각각 `0.5, 1, 2`일 때의 지수분포를 따르는 데이터를 `10000`개씩 뽑고 각 데이터셋의 밀도분포를 그리세요.



* 3개의 데이터셋을 비교하고, lambda 값에 따라 데이터가 어떻게 샘플링 되는지 간단히 설명하세요.

````python
ed_05 = np.random.exponential(scale=0.5 , size=10000)
ed_10 = np.random.exponential(scale=1.0 , size=10000)
ed_12 = np.random.exponential(scale=1.2 , size=10000)

plt.hist([ed_05,ed_10,ed_12] , histtype="step", bins=200, alpha=0.5, label=[0.5,1,2], linewidth=3)
plt.legend(fontsize=30, ncol=3)
plt.show()
````
