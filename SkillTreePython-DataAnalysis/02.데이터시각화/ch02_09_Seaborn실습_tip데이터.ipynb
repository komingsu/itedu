````python
df=sns.load_dataset('titanic')
df.info()
````

#### | 데이터 기본 탐색


````python
# age 데이터 fill
df["age"] = df["age"].fillna(df["age"].median())
````

````python
# embarked 데이터 fill
df["embarked"] = df["embarked"].fillna("S")
````

````python
# embarked_town 데이터 fill
df["embark_town"] = df["embark_town"].fillna("Southampton")
````

````python
# deck 데이터 fill -> N
df["deck"] = df["deck"].astype("object")
df["deck"] = df["deck"].fillna("N")
````

````python
# 가족 데이터 생성
df["family"] = df["sibsp"] + df["parch"]
````

````python
sns.catplot(data=df, x="alone", hue="sex", kind="count")
````
