````python
sns.relplot(data=fmri, x="timepoint", y="signal", kind="line", hue="event", size="region")
````

````python
sns.relplot(data=fmri, x="timepoint", y="signal", kind="line", hue="event", style="region")
````

````python
sns.relplot(data=fmri,
            x="timepoint", # x축 데이터
            y="signal", # y축 데이터
            kind="line", # 그래프 종류
            hue="event", # 색으로 나눌 칼럼
            style="region", # 선스타일로 나눌 칼럼
            col="subject", # 그래프를 나눌 칼럼
            col_wrap=4, # 그래프 묶음 개수
            linewidth=3, # 그래프 굵기
            col_order = [f"s{i}" for i in range(14)] # 그래프를 나타내는 순서
           )
````

````python
sns.relplot(data=fmri.query("region=='frontal'"), x="timepoint", y="signal", kind="line", hue="event", col="subject", col_wrap=4, linewidth=3,col_order = [f"s{i}" for i in range(14)])
````

````python
sns.catplot(data=fmri, x="timepoint", y="signal", kind="bar", hue="event", col="subject", col_wrap=4, ci=None, sharex=False, col_order = [f"s{i}" for i in range(14)])
````

### (3) Relational plot

````python
# sns.relplot(data=fmri,
#             x="timepoint", # x축 데이터
#             y="signal", # y축 데이터
#             kind="line", # 그래프 종류
#             hue="event", # 색으로 나눌 칼럼
#             style="region", # 선스타일로 나눌 칼럼
#             col="subject", # 그래프를 나눌 칼럼
#             col_wrap=4, # 그래프 묶음 개수
#             linewidth=3, # 그래프 굵기
#             col_order = [f"s{i}" for i in range(14)] # 그래프를 나타내는 순서
#            )

sns.relplot(data=df,
            x="month",
            y="passengers",
            hue="year",
            kind="line",
            linewidth=2,
            legend = "full",
            height=5, aspect=2,
            palette="rainbow")
````

````python
df.sort_values("year",ascending=False,inplace=True)
p2 = df.pivot_table(index="month", columns="year", values="passengers", sort=True)
p2.plot(figsize=(15,7))
plt.ylabel("passengers", labelpad=7, fontsize=20)
plt.xlabel("month", labelpad=7, fontsize=20)
plt.legend(title = "years",bbox_to_anchor=(0.6, 0.18, 0.5, 0.5))
plt.show()
````

````python
p2 = p2.T.iloc[::-1,:]
p2.plot(figsize=(15,7))
plt.ylabel("passengers", labelpad=7, fontsize=20)
plt.xlabel("month", labelpad=7, fontsize=20)
plt.legend(title = "years",bbox_to_anchor=(0.6, 0.18, 0.5, 0.5))
plt.show()
````
