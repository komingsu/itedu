````python
np.random.seed(42)
fruits=['apple', 'banana','orange']
fresh = np.random.randint(1,10,3)
qual = np.random.randint(1,10,3)
price = np.random.randint(1000,1500,3)
sale = np.random.randint(700,1200,3)

fresh, qual, price, sale
````

### (2) **kwargs 이용한 옵션적용




````python
idx = np.arange(3)/2
width=0.2

plt.figure(figsize=(10,8))
plt.subplot(3,1,1)
plt.bar(fruits, fresh, label="fresh", alpha=0.8)
plt.bar(fruits, qual, label="quality", alpha=0.8)
plt.legend()

plt.subplot(3,1,2)
plt.bar(idx, fresh, label="fresh", width=width)
plt.bar(idx+width, qual, label="quality", width=width)
plt.xticks(idx+(width/2), fruits)
plt.legend()

plt.subplot(3,1,3)
plt.bar(fruits, price, label="price")
plt.bar(fruits, sale, label="sale")
plt.legend()


plt.show()
````

````python
plt.figure(figsize=(10,7))

style = dict(width=width,
             alpha=0.5,
            linewidth=3,
            edgecolor="k")

plt.bar(idx, fresh, label="fresh",  **style)


plt.bar(idx+width, qual, label="quality", **style)
plt.xticks(idx+(width/2), fruits)
plt.legend()
````

### (1) 막대그래프를 이용한 flight 데이터 시각화

```
df=sns.load_dataset('flights')
```

 `seaborn`의 `load_dataset`을 이용해 `flights` 데이터를 불러온 후 아래와 같은 플롯을 그리세요.

![](https://github.com/yebiny/SkillTreePython-DataAnalysis/blob/main/imgs/ch0203-01.png?raw=true)





````python
df=sns.load_dataset('flights')

df_1950 = df[df["year"]==1950]
df_1955 = df[df["year"]==1955]
df_1960 = df[df["year"]==1960]
df_mean = pd.concat( [df_1950, df_1955, df_1960] ,  axis=0).groupby(["month"])["passengers"].mean()

idx = np.arange(12)/2
width=0.12

plt.bar(idx, df_1950.passengers, alpha=0.6, width=width, label="1950")
plt.bar(idx+width, df_1955.passengers, alpha=0.6, width=width, label="1955")
plt.bar(idx+width*2 , df_1960.passengers, alpha=0.6, width=width, label="1960")

plt.plot(idx+width , df_mean, label="mean of each year", marker="o", color="k", markerfacecolor="r")


plt.xticks(idx+(width), df_1950.month)
plt.legend(loc="upper left")

plt.show()
````
