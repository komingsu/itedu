### **01. 플롯 분할**
---

````python
x = np.linspace(-5,5, 1000)

def func2(a,b,c):
    y = a*(x-b)**2+c
    return y

plt.subplot(1, 2,1)
plt.plot(x, func2(2,4,-3))
plt.plot(x, func2(2,1,3))

plt.axvline(0,  color="k")
plt.axhline(0,  color="k")
plt.axis([-5,5,-1,100])


plt.subplot(1,2,2)
plt.plot(x, func2(1,0,0))
plt.plot(x, func2(3,1,0))

plt.axvline(0,  color="k")
plt.axhline(0,  color="k")
plt.axis([-5,5,-1,100])
````

### **02. 눈금 공유**
---



````python
t = np.linspace(0,2*np.pi,500)

s1 = np.sin(t)
s2 = np.sin(t**2)
s3 = np.exp(-t)

plt.plot(t,s1)
plt.plot(t,s2)
plt.plot(t,s3)
````

````python
t = np.linspace(0,2*np.pi,500)

s1 = np.sin(t)
s2 = np.sin(t**2)
s3 = np.exp(-t)

ax1 = plt.subplot(3,1,1)
plt.plot(t,s1)
plt.tick_params("x", labelbottom=False)

plt.xticks(np.linspace(0,2*np.pi,5), labels=["0","1/2 pi","pi","3/2 pi","2pi"])

plt.subplot(3,1,2, sharex=ax1)
plt.plot(t,s2)
plt.tick_params("x", labelbottom=False)

plt.subplot(3,1,3, sharex=ax1)
plt.plot(t,s3)

plt.show()
````

### **01. 로그함수와 지수함수 그리기**
---

로그함수

````python
import math

s = np.linspace(0.1,3,500)

# s1 = np.sin(s)

s1 = np.array(list(map(lambda x: math.log(x,2), s)))
s2 = np.array(list(map(lambda x: math.log(x,3), s)))
s3 = np.array(list(map(lambda x: math.log(x,0.5), s)))
s4 = np.array(list(map(lambda x: math.log(x,(1/3)), s)))
# math.log(8,2)
plt.axvline(0,  color="k", alpha=0.8)
plt.axhline(0,  color="k", alpha=0.8)

plt.plot(s,s1, label="log2(x)")
plt.plot(s,s2, label="log3(x)")
plt.plot(s,s3, label="log1/2(x)")
plt.plot(s,s4, label="log1/3(x)")
plt.scatter(1,0, linewidths=5, color="k")
plt.legend(bbox_to_anchor=(0.65, -0.3, 0.5, 0.5))
plt.show()
````

지수함수

````python
s = np.linspace(0,5,500)

s1 = np.array(list(map(lambda x: 2**x, s)))
s2 = np.array(list(map(lambda x: 3**x, s)))
s3 = np.array(list(map(lambda x: (1/2)**x, s)))
s4 = np.array(list(map(lambda x: (1/3)**x, s)))
s5 = np.exp(s)

plt.axvline(0,  color="k", alpha=0.8)
plt.axhline(0,  color="k", alpha=0.8)

plt.plot(s,s1, label="2^x")
plt.plot(s,s2, label="3^x")
plt.plot(s,s3, label="(1/2)^x")
plt.plot(s,s4, label="(1/3)^x")
plt.plot(s,s5, label="e^x")

plt.legend(bbox_to_anchor=(0.65, -0.25, 0.5, 0.5))
plt.show()
````

### filght 데이터 시각화

````python
df = sns.load_dataset("flights")
df.info()
````

````python
df.head()
````

````python
df.sort_values("year",ascending=False,inplace=True)
p2 = df.pivot_table(index="month", columns="year", values="passengers", sort=True)
p2.plot(figsize=(15,7))
plt.ylabel("passengers", labelpad=7, fontsize=20)
plt.xlabel("month", labelpad=7, fontsize=20)
plt.legend(title = "years",bbox_to_anchor=(0.58, 0, 0.5, 0.5))
plt.show()
````

````python
p2 = p2.T.iloc[::-1,:]
p2
p2.plot(figsize=(15,7))
plt.ylabel("passengers", labelpad=7, fontsize=20)
plt.xlabel("month", labelpad=7, fontsize=20)
plt.legend(title = "years",bbox_to_anchor=(0.58, 0, 0.5, 0.5))
plt.show()
````
