```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

train_gen = ImageDataGenerator(rescale=1.0/255.0).flow_from_directory("../../dataset/cats_and_dogs/new_ds/train",
                                                                      target_size=(150,150),
                                                                      batch_size = 20,
                                                                      class_mode = "binary") # "categorical", "binary", "sparse"
valid_gen = ImageDataGenerator(rescale=1.0/255.0).flow_from_directory("../../dataset/cats_and_dogs/new_ds/valid",
                                                                      target_size=(150,150),
                                                                      batch_size = 20,
                                                                      class_mode = "binary")
```

# 네트워크 구성하기

## 일반 데이터셋 사용

```python
def create_model():
    x = layers.Input(shape=(150,150,3))
    y = layers.Conv2D(32,3, activation="relu")(x)
    y = layers.MaxPool2D()(y)
    y = layers.Conv2D(64,3, activation="relu")(y)
    y = layers.MaxPool2D()(y)
    y = layers.Conv2D(128,3, activation="relu")(y)
    y = layers.MaxPool2D()(y)
    y = layers.Conv2D(128,3, activation="relu")(y)
    y = layers.MaxPool2D()(y)
    y = layers.Flatten()(y)
    y = layers.Dense(512, activation="relu")(y)
    y = layers.Dense(1, activation="sigmoid")(y)
    return models.Model(x,y)

model = create_model()
loss = losses.BinaryCrossentropy()
optimizer = optimizers.RMSprop(learning_rate=0.0001)

model.compile(loss=loss,
              optimizer=optimizer,
              metrics = ["acc"])
```

## ImageDataGenerater 사용

```python
aug_gen = ImageDataGenerator(rescale=1.0/255.0, # 스케일 조정
                             rotation_range=40, # 회전 범위
                             width_shift_range=0.2, # 좌우 움직이는 범위
                             height_shift_range=0.2, # 상하 움직이는 범위
                             shear_range=0.2, # 기울기 범위
                             zoom_range=0.2, # 확대 범위
                             horizontal_flip=True, # 좌우반전
                             fill_mode="nearest", # 부족한 부분을 채우는 방법
                             brightness_range = [0.8,1.2]) # 밝기 범위 (1 기준)
```

## 모델 생성

```python
def create_model2():
    x = layers.Input(shape=(150,150,3))
    y = layers.Conv2D(32,3, activation="relu")(x)
    y = layers.MaxPool2D()(y)
    y = layers.Conv2D(64,3, activation="relu")(y)
    y = layers.MaxPool2D()(y)
    y = layers.Conv2D(128,3, activation="relu")(y)
    y = layers.MaxPool2D()(y)
    y = layers.Conv2D(128,3, activation="relu")(y)
    y = layers.MaxPool2D()(y)
    y = layers.Flatten()(y)
    y = layers.Dropout(0.5)(y) # dropout 층을 추가
    y = layers.Dense(512, activation="relu")(y)
    y = layers.Dense(1, activation="sigmoid")(y)
    return models.Model(x,y)

model2 = create_model2()
model2.compile(loss=loss,
              optimizer=optimizer,
              metrics = ["acc"])
```
