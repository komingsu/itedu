```python
import matplotlib.pyplot as plt
import numpy as np
import os
import PIL
import tensorflow as tf
import pathlib

from tensorflow import keras
from tensorflow.keras import layers, models, utils, callbacks, losses, optimizers
from tensorflow.keras.models import Sequential

from Myfunc import show_acc_loss_hists
```

# 데이터셋

## 준비

```python
dataset_url = "https://storage.googleapis.com/download.tensorflow.org/example_images/flower_photos.tgz"
data_dir = tf.keras.utils.get_file('flower_photos', origin=dataset_url, untar=True)
data_dir = pathlib.Path(data_dir)
```

```python
data_dir
```

```python
list(data_dir.glob('*/*.jpg'))[:10]
```

## 데이터셋 설정

* subset 을 "training" 또는 "validation" 을 지정하고 동일한 seed 를 설정해야한다.

```python
batch_size = 32
img_height, img_width = 180, 180
```

```python
train_ds = tf.keras.preprocessing.image_dataset_from_directory( #폴더에서 이미지 바로 가져오기
    data_dir, # 폴더 path
    validation_split=0.2, # validation 으로 사용할 데이터 비율
    subset="training", # 데이터셋 이름
    seed=123, # 랜덤 seed
    image_size=(img_height, img_width), # 이미지 size
    batch_size=batch_size) # 가져올 batch size
```

```python
val_ds = tf.keras.preprocessing.image_dataset_from_directory(
  data_dir,
  validation_split=0.2,
  subset="validation",
  seed=123,
  image_size=(img_height, img_width),
  batch_size=batch_size)
```

```python
class_names = train_ds.class_names
print(class_names)
```

## 데이터셋 시각화 하기

```python
plt.figure(figsize=(10, 10))
for images, labels in train_ds.take(1):
    for i in range(9):
        ax = plt.subplot(3, 3, i + 1)
        plt.imshow(images[i].numpy().astype("uint8"))
        plt.title(class_names[labels[i]])
        plt.axis("off")
```

```python
for image_batch, labels_batch in train_ds:
    print(image_batch.shape)
    print(labels_batch.shape)
    break
```

## 데이터셋 성능강화

* `Dataset.cache()` : 첫 epoch 동안 디스크에서 이미지를 로드한 후 이미지를 메모리에 유지합니다. 
* `Dataset.prefetch()` : 훈련 중에 데이터 전처리 및 모델 실행과 겹칩니다.

```python
AUTOTUNE = tf.data.experimental.AUTOTUNE

train_ds = train_ds.cache().shuffle(1000).prefetch(buffer_size=AUTOTUNE)
val_ds = val_ds.cache().prefetch(buffer_size=AUTOTUNE)
```

# 모델 생성

## 일반 모델 생성

```python
num_classes = 5


def build_model():
    x = layers.Input(shape=(img_height, img_width, 3), name="Input")
    y = layers.experimental.preprocessing.Rescaling(1. / 255)(x)
    y = layers.Conv2D(16, 3, padding='same', activation='relu')(y)
    y = layers.MaxPooling2D()(y)
    y = layers.Conv2D(32, 3, padding='same', activation='relu')(y)
    y = layers.MaxPooling2D()(y)
    y = layers.Conv2D(64, 3, padding='same', activation='relu')(y)
    y = layers.MaxPooling2D()(y)
    y = layers.Flatten()(y)
    y = layers.Dense(128, activation='relu')(y)
    y = layers.Dense(num_classes)(y)

    return models.Model(x, y)


model = build_model()
model.compile(
    optimizer='adam',
    loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
    metrics=['acc'])
```

## 데이터셋 증강 모델 생성

```python
def data_augmentation_module(x):
    y = layers.experimental.preprocessing.RandomFlip("horizontal", input_shape=(img_height, img_width,3))(x)
    y = layers.experimental.preprocessing.RandomRotation(0.1)(y)
    y = layers.experimental.preprocessing.RandomZoom(0.1)(y)
    return y
```

```python
def build_model2():
    x = layers.Input(shape=(img_height, img_width, 3), name="Input")
    y = data_augmentation_module(x)
    y = layers.experimental.preprocessing.Rescaling(1. / 255)(y)
    y = layers.Conv2D(16, 3, padding='same', activation='relu')(y)
    y = layers.MaxPooling2D()(y)
    y = layers.Conv2D(32, 3, padding='same', activation='relu')(y)
    y = layers.MaxPooling2D()(y)
    y = layers.Conv2D(64, 3, padding='same', activation='relu')(y)
    y = layers.MaxPooling2D()(y)
    y = layers.Dropout(0.2)(y)
    y = layers.Flatten()(y)
    y = layers.Dense(128, activation='relu')(y)
    y = layers.Dense(num_classes)(y)
    return models.Model(x, y)


model2 = build_model2()
model2.compile(
    optimizer='adam',
    loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
    metrics=['acc'])
```

## 콜백

```python
ckp1 = callbacks.ModelCheckpoint("model_ex_1.h5", save_best_only=True, monitor="val_loss")
ckp2 = callbacks.ModelCheckpoint("model_ex_2.h5", save_best_only=True, monitor="val_loss")
early_stopping = callbacks.EarlyStopping(monitor="val_loss", patience=7)
```

## 훈련

### org model 훈련

```python
history = model.fit(train_ds,
                    epochs=50,
                    validation_data=val_ds,
                    verbose=1,
                    callbacks=[ckp1, early_stopping])
```

```python
show_acc_loss_hist(history)
```

### 데이터셋 증강 모델 훈련

```python
history2 = model2.fit(train_ds,
                    epochs=50,
                    validation_data=val_ds,
                    verbose=1,
                    callbacks=[ckp1, early_stopping])
```

```python
show_acc_loss_hist(history2)
```
