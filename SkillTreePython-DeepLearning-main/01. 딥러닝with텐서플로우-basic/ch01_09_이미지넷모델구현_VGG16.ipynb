## 소개
---

```python
import tensorflow as tf
from tensorflow.keras import models, layers, utils
```

## 구현
---

```python
def VGG_16():
    x = layers.Input(shape=(224,224,3))
    
    y = layers.Conv2D(filters=64, kernel_size=(3,3), padding="same", activation="relu")(x)
    y = layers.Conv2D(filters=64,kernel_size=(3,3),padding="same", activation="relu")(y)
    y = layers.MaxPool2D(pool_size=(2,2), strides=(2,2))(y)
    
    y = layers.Conv2D(filters=128, kernel_size=(3,3), padding="same", activation="relu")(y)
    y = layers.Conv2D(filters=128,kernel_size=(3,3),padding="same", activation="relu")(y)
    y = layers.MaxPool2D(pool_size=(2,2), strides=(2,2))(y)
    
    y = layers.Conv2D(filters=256, kernel_size=(3,3), padding="same", activation="relu")(y)
    y = layers.Conv2D(filters=256, kernel_size=(3,3), padding="same", activation="relu")(y)
    y = layers.Conv2D(filters=256, kernel_size=(3,3), padding="same", activation="relu")(y)
    y = layers.MaxPool2D(pool_size=(2,2), strides=(2,2))(y)

    y = layers.Conv2D(filters=512, kernel_size=(3,3), padding="same", activation="relu")(y)
    y = layers.Conv2D(filters=512, kernel_size=(3,3), padding="same", activation="relu")(y)
    y = layers.Conv2D(filters=512, kernel_size=(3,3), padding="same", activation="relu")(y)
    y = layers.MaxPool2D(pool_size=(2,2), strides=(2,2))(y)
    
    y = layers.Conv2D(filters=512, kernel_size=(3,3), padding="same", activation="relu")(y)
    y = layers.Conv2D(filters=512, kernel_size=(3,3), padding="same", activation="relu")(y)
    y = layers.Conv2D(filters=512, kernel_size=(3,3), padding="same", activation="relu")(y)
    y = layers.MaxPool2D(pool_size=(2,2), strides=(2,2))(y)
    
    y = layers.Flatten()(y)
    y = layers.Dense(units=4096, activation="relu")(y)
    y = layers.Dense(units=4096, activation="relu")(y)
    y = layers.Dense(units=2, activation="softmax")(y)
    
    return y
```
