```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
cv2.__version__
```

# 1번

```python
shape = (100, 100, 3)
value = 255
bkg = np.full(shape,value, np.uint8)
cv2.circle(bkg, (50, 50), 30, (0,255,0), 2)
plt.grid()
_ = plt.imshow(bkg)
```

# 3번

```python
line = np.linspace(0,255,200, dtype=np.uint8).reshape(200,1)
img2 = np.concatenate([line for i in range(200)],axis=1)
print(img2.shape)

cv2.imshow("img2", img2)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
ret, dst = cv2.threshold(img2, 123,255, cv2.THRESH_BINARY)
cv2.imshow("dst",dst)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# 7번

```python
bkg = np.zeros((100,100), dtype = "uint8")
cv2.circle(bkg, (50, 50), 30, (255,255,255), -1)
plt.grid()
_ = plt.imshow(bkg, cmap='gray', vmin=0, vmax=255)
```

```python
mask = np.zeros((100,100), dtype = "uint8")
mask = cv2.rectangle(mask, (50, 80), (80,20), (255,255,255), -1)
mask = cv2.rectangle(mask, (20, 80), (50,50), (255,255,255), -1)
plt.grid()
_ = plt.imshow(mask, cmap='gray', vmin=0, vmax=255)
```

```python
masked = cv2.bitwise_and(bkg, bkg, mask=mask)
plt.grid()
_ = plt.imshow(masked, cmap='gray', vmin=0, vmax=255)
```
