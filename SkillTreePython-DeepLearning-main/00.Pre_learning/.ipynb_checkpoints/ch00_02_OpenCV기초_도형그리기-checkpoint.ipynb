### **| 색깔 파라미터 설정**

````python
def nothing(x):
    pass

img = np.zeros((300,512,3), np.uint8)
cv2.namedWindow('image')

# trackbar를 생성하여 named window에 등록
cv2.createTrackbar('R', 'image', 0, 255, nothing)
cv2.createTrackbar('G', 'image', 0, 255, nothing)
cv2.createTrackbar('B', 'image', 0, 255, nothing)


while(1):
    cv2.imshow('image', img)

    if cv2.waitKey(1) & 0xFF == 27:
        break

    r = cv2.getTrackbarPos('R','image')
    g = cv2.getTrackbarPos('G', 'image')
    b = cv2.getTrackbarPos('B', 'image')

    
    img[:] = [b,g,r] # 모든 행/열 좌표값을 [b,g,r]로 변경

cv2.destroyAllWindows()
````

````python
R, G, B = 90, 30, 90
COLOR = (B,G,R)
````

## 도형 그리기

### **| Line 그리기**

Start와 End 점을 연결하여 직선을 그립니다.

#### `cv2.line(img, start, end, color, thickness)`
* Parameters
  * img – 그림을 그릴 이미지 파일
  * start – 시작 좌표(ex; (0,0))
  * end – 종료 좌표(ex; (500. 500))
  * color – BGR형태의 Color(ex; (255, 0, 0) -> Blue)
  * thickness (int) – 선의 두께. pixel

````python
leo_copy = leo.copy()
cv2.line(leo_copy, (100,120), (310,320), COLOR, 3)
cv2.line(leo_copy, (310,120), (100,320), COLOR, 3)
cv2.imshow("img", leo_copy)
cv2.waitKey(0)
cv2.destroyAllWindows()b
````

### **| 사각형 그리기**

top-left corner와 bottom-right corner점을 연결하는 사각형을 그립니다.

#### `cv2.rectangle(img, start, end, color, thickness)`
* Parameters
  * img – 그림을 그릴 이미지
  * start – 시작 좌표(ex; (0,0))
  * end – 종료 좌표(ex; (500. 500))
  * color – BGR형태의 Color(ex; (255, 0, 0) -> Blue)
  * thickness (int) – 선의 두께. pixel

* Sample Code
```
img = cv2.rectangle(img, (384, 0), (510, 128), (0,255,0), 3)
```

````python
leo_copy = leo.copy()
img = cv2.rectangle(leo_copy, (100, 120), (310,320), COLOR, 3)
cv2.imshow("img", img)
cv2.waitKey(0)
cv2.destroyAllWindows()
````

### **| 원 그리기**

top-left corner와 bottom-right corner점을 연결하는 사각형을 그립니다.

#### `cv2.rectangle(img, start, end, color, thickness)`
* Parameters
  * img – 그림을 그릴 이미지
  * center – 원의 중심 좌표(x, y)
  * radian – 반지름
  * color – BGR형태의 Color
  * thickness – 선의 두께, -1 이면 원 안쪽을 채움

* Sample Code
```
img = cv2.circle(img, (447,63), 63, (0,0,255), -1)
```

````python
leo_copy = leo.copy()
img = cv2.circle(leo_copy, (210, 230), 100, COLOR, 3)
cv2.imshow("img", img)
cv2.waitKey(0)
cv2.destroyAllWindows()
````

### **| 텍스트 추가**

#### `cv2.putText(img, text, org, font, fontSacle, color, thickness)`
* Parameters
 * img – image
 * text – 표시할 문자열
 * org – 문자열이 표시될 위치. 문자열의 bottom-left corner점
 * font – font type. CV2.FONT_XXX
 * fontSacle – Font Size
 * color – fond color
 * thickness - 선의 두께

* Sample Code
```
cv2.putText(img, 'OpenCV', (10,500), cv2.FONT_HERSHEY_SIMPLEX, 4, (255,255,255), 2)
```


### OpenCV 에서 사용하는 글꼴 종류

 1. cv2.FONT_HERSHEY_SIMPLEX : 보통 크기의 산 세리프 글꼴
 1. cv2.FONT_HERSHEY_PLAIN : 작은 크기의 산 세리프 글꼴
 1. cv2.FONT_HERSHEY_SCRIPT_SIMPLEX : 필기체 스타일 글꼴
 1. cv2.FONT_HERSHEY_TRIPLEX : 보통크기의 세리프 글꼴
 1. cv2.FONT_ITALIC : 기울임(이탤릭체)

````python
leo_copy = leo.copy()
img = cv2.putText(leo_copy, 'LEO', (100, 130), cv2.FONT_HERSHEY_SIMPLEX, 4, COLOR, 2)
cv2.imshow("img", img)
cv2.waitKey(0)
cv2.destroyAllWindows()
````

#### **예제 01**

OpenCV를 이용해 이미지에 아래와 같이 도형을 그려보세요

![](https://github.com/yebiny/SkillTreePython-DeepLearning/blob/main/imgs/cv2_1.png?raw=true)

````python
shape = (100, 100, 3)
value = 255
bkg = np.full(shape,value, np.uint8)
plt.imshow(bkg)

cv2.rectangle(bkg, (10,50), (20,60), COLOR, 2)
cv2.line(bkg,(30,50), (40,60), COLOR, 2)
cv2.circle(bkg, (50, 55), 5, COLOR, 2)
cv2.putText(bkg, 'Hello', (60, 60), cv2.FONT_HERSHEY_SIMPLEX, 0.4, COLOR, 1)

plt.imshow(bkg)
````

#### **예제 02**

OpenCV를 이용해 이미지에 아래와 같이 도형을 그려보세요


![](https://github.com/yebiny/SkillTreePython-DeepLearning/blob/main/imgs/cv2_2.png?raw=true)

````python
shape = (100, 100, 3)
value = 255
bkg = np.full(shape,value, np.uint8)
plt.imshow(bkg)

for i in [0,20,40,60,80]:
    cv2.line(bkg,(i,i), (i,i+20), COLOR,2)
    cv2.line(bkg,(i,i), (i+20,i), COLOR, 2)
    cv2.line(bkg,(i+20,i), (i+20,i+20), COLOR, 2)
    cv2.line(bkg,(i,i+20), (i+20,i+20), COLOR, 2)

cv2.circle(bkg, (20, 80), 20, COLOR, -1)    
cv2.circle(bkg, (80, 20), 20, COLOR, -1) 
plt.imshow(bkg)
````
