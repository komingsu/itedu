## imdb 베이스 모델
---

```python
import tensorflow as tf
from tensorflow.keras import preprocessing, layers, Sequential
from Myfunc import show_lcurve
```

```python
MAX_LEN = 256
MAX_FEATURES = 10000

imdb = tf.keras.datasets.imdb
(x_train, y_train), (x_test, y_test) = imdb.load_data(num_words=MAX_FEATURES)
x_train = preprocessing.sequence.pad_sequences(x_train,
                                               padding='post',
                                               maxlen=MAX_LEN)

x_test = preprocessing.sequence.pad_sequences(x_test,
                                              padding='post',
                                              maxlen=MAX_LEN)
print(x_train.shape, x_test.shape)
```

### **| FCN with GlobalAveragePooling1D**

```python
EMD_DIMS = 32
model = Sequential(name="GAP")
model.add(layers.Input(shape=MAX_LEN, name="Input"))
model.add(layers.Embedding(input_dim=10000, output_dim=32, name="Embedding"))
model.add(layers.GlobalAveragePooling1D()) # channels_last 가 디폴트 , # channels_last 를 해주면 각 행별 평균을 한다
model.add(layers.Dense(1, activation="sigmoid", name="Classification"))
model.summary()

model.compile(optimizer="rmsprop", loss="binary_crossentropy", metrics=["acc"])
```

```python
history_fcn = model.fit(x_train,y_train,
                    epochs=20,
                    batch_size=128,
                    validation_split=0.2,
                    verbose=1)
```

```python
show_lcurve([history_fcn], ["fcn"], ["r"])
```

## RNN
---

<p align='center'>
<img src=https://res.cloudinary.com/dyd911kmh/image/upload/v1647442110/image6_f6vds6.png width=800>
</p>

```python
EMD_DIMS = 32
model = Sequential(name="RNN")
model.add(layers.Input(shape=MAX_LEN, name="Input"))
model.add(layers.Embedding(input_dim=10000, output_dim=32, name="Embedding"))
model.add(layers.SimpleRNN(30, name="RNN")) # channels_last 가 디폴트 , # channels_last 를 해주면 각 행별 평균을 한다
model.add(layers.Dense(1, activation="sigmoid", name="classifier"))
model.summary()

model.compile(optimizer="rmsprop", loss="binary_crossentropy", metrics=["acc"])
```

* RNN 의 파라미터 개수 : (32 * 30)+(30 * 30) + (1 * 30) = 1890
* (RNN * RNN) + (Input_dim * RNN) + (RNN_dim * RNN)

```python
history_rnn = model.fit(x_train,y_train,
                    epochs=20,
                    batch_size=128,
                    validation_split=0.2,
                    verbose=1)
```

```python
show_lcurve([history_rnn, history_fcn],["rnn","fcn"],["r","b"])
```

## LSTM
---

<p align='center'>
<img src=https://res.cloudinary.com/dyd911kmh/image/upload/v1647442110/image3_nsnxp3.png width=600>
</p>



```python
EMD_DIMS = 32
model = Sequential(name="LSTM")
model.add(layers.Input(shape=MAX_LEN, name="Input"))
model.add(layers.Embedding(input_dim=10000, output_dim=32, name="Embedding"))
model.add(layers.LSTM(30, name="LSTM")) # channels_last 가 디폴트 , # channels_last 를 해주면 각 행별 평균을 한다
model.add(layers.Dense(1, activation="sigmoid", name="classifier"))
model.summary()

model.compile(optimizer="rmsprop", loss="binary_crossentropy", metrics=["acc"])
```

* 파라미터 개수 : 4 * (33 * 30 + 30 * 30) = 7560
* params = 4 * ((size_of_input + 1) * size_of_output + size_of_output^2)

```python
history_lstm = model.fit(x_train,y_train,
                         epochs=20,
                         batch_size=128,
                         validation_split=0.2,
                         verbose=1)
```

```python
show_lcurve([history_fcn, history_lstm],["gap","lstm"],["r","b"], size=(18,8))
```

### | **many-to-many**


<p align='center'>
<img src=https://res.cloudinary.com/dyd911kmh/image/upload/v1647442111/image9_e1gwba.png width=700>
</p>


```python
EMD_DIMS = 32
model = Sequential(name="LSTM_mtm")
model.add(layers.Input(shape=MAX_LEN, name="Input"))
model.add(layers.Embedding(input_dim=10000, output_dim=32, name="Embedding"))
model.add(layers.LSTM(32, return_sequences=True, name="LSTM_1"))
model.add(layers.LSTM(32, return_sequences=False, name="LSTM_2"))
# return_sequences=True 이면 output이 2차텐서 (,256,32)
# return_sequences=False 이면 output이 1차텐서 (,32)
model.add(layers.Dense(1, activation="sigmoid", name="classifier"))
model.summary()

model.compile(optimizer="rmsprop", loss="binary_crossentropy", metrics=["acc"])
```

```python
history_lstm_mtm = model.fit(x_train, y_train,
                             epochs=20,
                             batch_size=128,
                             validation_split=0.2,
                             verbose=1)
```

```python
show_lcurve([history_lstm, history_lstm_mtm],["lstm","lstm_mtm"],["r","b"], size=(18,8))
```

## GRU
---


<p align='center'>
<img src=https://res.cloudinary.com/dyd911kmh/image/upload/v1647442111/image10_zgefie.png width=600>
</p>



```python
EMD_DIMS = 32
model = Sequential(name="GRU")
model.add(layers.Input(shape=MAX_LEN, name="Input"))
model.add(layers.Embedding(input_dim=10000, output_dim=EMD_DIMS, name="Embedding"))
model.add(layers.GRU(30, name="GRU"))
model.add(layers.Dense(1, activation="sigmoid", name="classifier"))
model.summary()

model.compile(optimizer="rmsprop", loss="binary_crossentropy", metrics=["acc"])
```

```python
history_gru = model.fit(x_train, y_train,
                             epochs=20,
                             batch_size=128,
                             validation_split=0.2,
                             verbose=1)
```

```python
show_lcurve([history_fcn, history_lstm, history_gru],["gap","lstm","gru"],["r","b","g"], size=(18,8))
```

### | **many-to-many**


<p align='center'>
<img src=https://res.cloudinary.com/dyd911kmh/image/upload/v1647442111/image9_e1gwba.png width=700>
</p>


```python
EMD_DIMS = 32
model = Sequential(name="GRU")
model.add(layers.Input(shape=MAX_LEN, name="Input"))
model.add(layers.Embedding(input_dim=10000, output_dim=32, name="Embedding"))
model.add(layers.GRU(32, return_sequences=True, name="GRU_1"))
model.add(layers.GRU(32, name="GRU_2"))
model.add(layers.Dense(1, activation="sigmoid", name="classifier"))
model.summary()

model.compile(optimizer="rmsprop", loss="binary_crossentropy", metrics=["acc"])
```

```python
history_gru_mtm = model.fit(x_train, y_train,
                            epochs=20,
                            batch_size=128,
                            validation_split=0.2,
                            verbose=1)
```

```python
show_lcurve([history_fcn, history_gru, history_gru_mtm],["gap","gru","gru_mtm"],["r","b","g"], size=(18,8))
```
